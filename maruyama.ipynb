# 単振動進行波

```python
v = 3 # m/s
f = 10 # Hz
duration = 9 # s

w = 2*np.pi*f # [rad]
T = 1/f # [s]

k = w/v

phase_lower, phase_upper = [w * d for d in [0, duration]]

fs = 1000 # [Hz]
xs = 100 # [/m]
t = np.arange(0,9,1/fs)
x_min, x_max = [0, 10]
x = np.arange(x_min,x_max,1/xs)

images = []
for tau in t:
    p_tau = list()
    for i in x:
        theta = w*tau-k*i
        if phase_lower <= theta <= phase_upper:
            p_tau.append(np.cos(theta))
        else:
            p_tau.append(0)
    p_tau = np.array(p_tau)
    wav1=ax1.plot(x,p_tau,color="red",linestyle="None",markersize="3",marker="o")
    images.append(wav1)

#アニメーションを作成
fig
ani=animation.ArtistAnimation(fig,images,interval=100)
plt.close()
rc('animation', html='jshtml')
ani

```
